```python
# import dependencies 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load streaming data

df = pd.read_csv('./spotify.csv', index_col='Date', parse_dates=True)
df
```

```python
# ploting data

# set figure size
plt.figure(figsize=(16,7))

# set the title of the chart
plt.title("Daily Global Streams of 5 Popular Songs in 2017-2018")

# plot streaming data into linechart
sns.lineplot(data=df)
```

```python
# list all columns
list(df.columns)
```

```python
# plot subset of data

plt.figure(figsize=(14,6))
plt.title("Daily Global Streams of 2 Most Popular Songs in 2017-2018")
plt.xlabel("Date")

sns.lineplot(data=df['Shape of You'], label='Shape of You')
sns.lineplot(data=df['Despacito'], label='Despacito')
```
