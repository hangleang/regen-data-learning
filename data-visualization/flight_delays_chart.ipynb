```python
# import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("setup complete")
```

```python
# load flight data

df = pd.read_csv('./flight_delays.csv', index_col='Month')
df
```

```python
# plot data into bar chart

plt.figure(figsize=(16, 8))
plt.title("Average monthly arrival delays")

sns.barplot(x=df.index, y=df['AS'])
plt.ylabel("Arrival delays (in minutes)")
```

```python
# plot flight data into heatmap

plt.figure(figsize=(14,7))
plt.title("Average monthly arrival delays by each airline")

sns.heatmap(data=df, annot=True)
plt.xlabel("Airline")
```

the months toward end of the year, especially 9-11, are relatively dark, means the arrival of the flights during these months is good at keeping schedule
