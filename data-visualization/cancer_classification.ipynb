```python
# import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("setup complete")
```

```python
df = pd.read_csv('./cancer.csv', index_col='Id')
df.head()
```

```python
# Histograms for benign and maligant tumors

plt.title('Counts to compare between benign(B) and malignant(M) by area')
sns.histplot(data=df, x='Area (mean)', hue='Diagnosis')
```

```python
# KDE plots for benign and malignant tumors

plt.title('Compare between benign(B) and malignant(M) by radius')
sns.kdeplot(data=df, x='Radius (worst)', hue='Diagnosis', shade=True)
```
