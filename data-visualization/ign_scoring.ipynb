```python
# import dependencies

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("setup complete");
```

```python
# load IGN data
df = pd.read_csv('./ign_scores.csv', index_col='Platform')
df
```

```python
# plot data into bar chart

plt.figure(figsize=(10,7))
plt.title("Average score for racing games by platforms")

sns.barplot(x=df['Racing'], y=df.index)
plt.xlabel("Average score")
```

```python
# plot data into heat map

plt.figure(figsize=(10,10))
plt.title("Average score by platform and genre")

sns.heatmap(data=df, annot=True)
```
