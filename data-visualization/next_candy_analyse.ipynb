```python
# import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("setup complete")
```

```python
# load dataset

df = pd.read_csv('./candy.csv', index_col='id')
df.head()
```

```python
# plot data into scatter plot

plt.figure(figsize=(10,10))
sns.scatterplot(x=df['sugarpercent'], y=df['winpercent'])
plt.xlabel('Sugar Content')
plt.ylabel('Winning Rate')
```

```python
# plot data into regression plot

plt.figure(figsize=(10,10))
sns.regplot(x=df['sugarpercent'], y=df['winpercent'])
plt.xlabel('Sugar Content')
plt.ylabel('Winning Rate')
```

```python
# plot pricing and winning rate date into scatter plot chart

plt.figure(figsize=(10,10))
sns.scatterplot(x=df['pricepercent'], y=df['winpercent'], hue=df['chocolate'])
plt.xlabel('Pricing')
plt.ylabel('Winning Rate')
```

```python
# plot data to compare between chocolate candy and candy without chocolate
sns.lmplot(x='pricepercent', y='winpercent', hue='chocolate', data=df)
plt.xlabel('Pricing')
plt.ylabel('Winning Rate')
```

```python
# plot data info categorical scatter plot
sns.swarmplot(x=df['chocolate'], y=df['winpercent'])
```
