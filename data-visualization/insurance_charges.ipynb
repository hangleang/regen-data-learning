```python
# import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("setup complete")
```

```python
# load insurance data

df = pd.read_csv('./insurance.csv')
df.head()
```

```python
# plot data into scatter plot

plt.figure(figsize=(7,7))
plt.title("Insurances Charges")
sns.scatterplot(x=df['bmi'], y=df['charges'], hue=df['smoker'])
```

```python
# plot data into chart with regression line added

plt.figure(figsize=(7,7))
plt.title("Insurances Charges")
sns.regplot(x=df['bmi'], y=df['charges'])
```

```python
# plot data into chart with two regression lines added

sns.lmplot(data=df, x='bmi', y='charges', hue='smoker')
```

```python
# plot data into categorical scatter plot

sns.swarmplot(x=df['smoker'], y=df['charges'])
```
