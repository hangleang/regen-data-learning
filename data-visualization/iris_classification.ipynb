```python
# import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("setup complete")
```

```python
# load plants data

df = pd.read_csv('./iris.csv', index_col='Id')
df.head()
```

```python
# plot data into histogram chart for counting identical values

# simple histogram
# sns.histplot(df['Petal Length (cm)'])

# categorical histogram chart with single dimension
sns.histplot(data=df, x='Petal Length (cm)', hue=df['Species'])

# categorical histogram chart with 2-dimensional fields
# sns.histplot(data=df, x='Petal Length (cm)', y='Sepal Width (cm)', hue=df['Species'])
```

```python
# plot data into KDE plot for density analysis (Kernel Density Estimate)

# simple kde plot
# sns.kdeplot(data=df['Petal Length (cm)'], shade=True)

# categorical kde plot with single dimension field
sns.kdeplot(data=df, x='Petal Length (cm)', hue='Species', shade=True)

# categorical kde plot with 2 dimensions field
# sns.kdeplot(data=df, x='Petal Length (cm)', y='Sepal Width (cm)', hue='Species', shade=True)
```

```python
# plot data into 2D KDE plot using `jointplot`

sns.jointplot(x=df['Petal Length (cm)'], y=df['Sepal Width (cm)'], hue=df['Species'], kind='kde')
```

### Conclusion

based on all of above charts, we can see that `iris-versicolor` and `iris-virginica` have some similar properties, not only their scientific name, but also its characteristics (Width, Length). we also see that only `iris-setosa` has its own characteristics, which is not identical with both above species.
