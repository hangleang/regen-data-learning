```python
# import dependencies 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("setup complete")
```

```python
# load data
df = pd.read_csv('./museum_visitors.csv', index_col='Date', parse_dates=True)
```

```python
# plotting data into linechart
plt.figure(figsize=(16,8))
plt.title("number of monthly visitors in 2017-2018")

sns.lineplot(data=df)
```

```python
# Assess seasonality
plt.figure(figsize=(16,8))
plt.title("number of monthly visitors in Avila Adobe Museum in 2017-2018")

sns.lineplot(data=df['Avila Adobe'], label='Avila Adobe')
```

### Conclusion
According to the above chart, we can see the pattern of decreasing visitors usually happen during Fall & Winter seasons, and start increasing gradually after that. In other words, the number of visitors start increasing from Sprint season, all the way up and peek during the Summer season.  The pattern involved over time as you can see in the chart, so we suggest to hire more seasonal staff during Spring & Summer seasons.
